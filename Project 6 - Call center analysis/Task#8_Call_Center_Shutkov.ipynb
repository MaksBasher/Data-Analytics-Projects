```python
import pandas as pd
import numpy as np
```

```python
call_center_data = pd.read_csv("Call_Center_Data.csv")
call_center_data
```

```python
call_center_data['Abandoned Calls'] = call_center_data['Incoming Calls'] - call_center_data['Answered Calls']
call_center_data
```

```python
call_center_data["Answering Rate,%"] =(call_center_data['Answered Calls']*100)/ call_center_data["Incoming Calls"]
pd.set_option('display.float_format', lambda x: '%.2f' % x)
call_center_data
```

```python
call_center_data["Abandonment Rate,%"] = 100 - (call_center_data['Answering Rate,%'])
pd.set_option('display.float_format', lambda x: '%.2f' % x)
call_center_data
```

```python
call_center_data['Answer_Speed_AVG'] = pd.to_datetime(call_center_data.Answer_Speed_AVG)
call_center_data['Answer_Speed_SEC'] = (pd.to_timedelta(call_center_data['Answer_Speed_AVG'].dt.strftime('%H:%M:%S')).dt.total_seconds().astype(int))
call_center_data
```

```python
#making classes
answering_speed = [
    (call_center_data['Answer_Speed_SEC'] <= 10),
    (call_center_data['Answer_Speed_SEC'] > 10) &  (call_center_data['Answer_Speed_SEC'] <= 20),
    (call_center_data['Answer_Speed_SEC'] > 20)
]
values = ['FAST', 'MEDIUM', 'SLOW']
call_center_data["Answering_Speed"] = np.select(answering_speed,values)
call_center_data
```

```python
#How many employees have a Telephone Answering Rate of more than 80%
answer_1 = len(call_center_data[call_center_data['Answering Rate,%']>80])
print(answer_1)
#Based on the new employee call answering speed category; what percentage of employees fall in the category of MEDIUM?
answer_2 = ((len(call_center_data[call_center_data['Answering_Speed']=="MEDIUM"])))*100/1251
print(answer_2, "%")
#How many employees receive less than 50 calls?
answer_3 = len(call_center_data[call_center_data['Incoming Calls']<50])
print(answer_3)
#What percentage of employees who have a Talk Duration of more than 2 minutes (120 seconds)?
call_center_data['Talk_Duration_AVG'] = pd.to_datetime(call_center_data.Talk_Duration_AVG)
call_center_data['Talk_Duration_SEC'] = (pd.to_timedelta(call_center_data['Talk_Duration_AVG'].dt.strftime('%H:%M:%S')).dt.total_seconds().astype(int))
answer_4 = ((len(call_center_data[call_center_data['Talk_Duration_SEC']>120]))*100)/1251
print(answer_4,"%")
```
